1. Считать изображение в формате RGB, перевести его в указанную в варианте цветовую модель.

```python
M = np.array([[0.412453, 0.357580, 0.180423],
        [0.212671, 0.715160, 0.072169],
        [0.019334, 0.119193, 0.950227]])
# region 辅助函数
# RGB2XYZ空间的系数矩阵
```

```python
# im_channel取值范围：[0,1]
def f(im_channel):
    return np.power(im_channel, 1 / 3) if im_channel > 0.008856 else 7.787 * im_channel + 0.137931

def anti_f(im_channel):
    return np.power(im_channel, 3) if im_channel > 0.206893 else (im_channel - 0.137931) / 7.787
# endregion
```

```python
# region RGB 转 Lab
# 像素值RGB转XYZ空间，pixel格式:(B,G,R)
# 返回XYZ空间下的值
def __rgb2xyz__(pixel):
    b, g, r = pixel[0], pixel[1], pixel[2]
    rgb = np.array([r, g, b])
    # rgb = rgb / 255.0
    # RGB = np.array([gamma(c) for c in rgb])
    XYZ = np.dot(M, rgb.T)#矩阵相乘，rgb.T转置
    XYZ = XYZ / 255.0
    return (XYZ[0] / 0.95047, XYZ[1] / 1.0, XYZ[2] / 1.08883)

def __xyz2lab__(xyz):
    """
    XYZ空间转Lab空间
    :param xyz: 像素xyz空间下的值
    :return: 返回Lab空间下的值
    """
    F_XYZ = [f(x) for x in xyz]
    L = 116 * F_XYZ[1] - 16 if xyz[1] > 0.008856 else 903.3 * xyz[1]
    a = 500 * (F_XYZ[0] - F_XYZ[1])
    b = 200 * (F_XYZ[1] - F_XYZ[2])
    return (L, a, b)

def _RGB2Lab_(pixel):
    """
    RGB空间转Lab空间
    :param pixel: RGB空间像素值，格式：[G,B,R]
    :return: 返回Lab空间下的值
    """
    xyz = __rgb2xyz__(pixel)
    Lab = __xyz2lab__(xyz)
    return Lab





 
```

```python
def _lab2xyz_(Lab):
    fY = (Lab[0] + 16.0) / 116.0
    fX = Lab[1] / 500.0 + fY
    fZ = fY - Lab[2] / 200.0

    x = anti_f(fX)
    y = anti_f(fY)
    z = anti_f(fZ)

    x = x * 0.95047
    y = y * 1.0
    z = z * 1.0883
    return(x,y,z)

def _xyz2rgb_(xyz):
    xyz = np.array(xyz)
    xyz = xyz * 255
    rgb = np.dot(np.linalg.inv(M), xyz.T)#Обратная матрица M
    rgb = np.uint8(np.clip(rgb, 0, 255))
    return rgb

def _Lab2RGB_(Lab):
    xyz = _lab2xyz_(Lab)
    rgb = _xyz2rgb_(xyz)
    return rgb

  
```

```python
from google.colab.patches import cv2_imshow
from IPython.display import display, HTML

w = img.shape[0]
h = img.shape[1]
img_new = np.zeros((w,h,3))
lab = np.zeros((w,h,3))
lab_l = np.zeros((w,h,3))
#rgb -> lab
for i in range(w):
  for j in range(h):
    Lab = _RGB2Lab_(img[i,j])
    lab[i, j] = (Lab[0], Lab[1], Lab[2])
#lab -> rgb
for i in range(w):
  for j in range(h):
    rgb = _Lab2RGB_(lab[i,j])
    img_new[i, j] = (rgb[2], rgb[1], rgb[0])
both = np.hstack((img, lab, img_new))
display(HTML("<font color='black'>|------------------ Исходное------------------|----------------Lab----------------|----------RGB ->----Lab ->  RGB----------|</font>"))
cv2_imshow(both)
```

```python
from IPython.display import display, HTML
from ipywidgets import FloatSlider

lab_L = np.zeros((w,h,3))
lab_a = np.zeros((w,h,3))
lab_b = np.zeros((w,h,3))
fSlider = FloatSlider(
    value=0, # Первоначальное значение
    min=-255,     # Минимум
    max=255,  # Максимум
    step=5,  # Шаг изменения
    description='Параметр:',
    continuous_update=False,  # True - событие observe возникает для каждого шага при изменении значения
    orientation='vertical'  # Горизонтальное или вертикальное расположение
)

def on_value_change(b):
    print(b['old'], '->', b['new'])
    value = b['new']
    for i in range(w):
      for j in range(h):
        rgb = _Lab2RGB_(lab[i,j])
        img_new[i, j] = (rgb[2]+value, rgb[1]+value, rgb[0]+value)

    for i in range(w):
      for j in range(h):
        Lab = _RGB2Lab_(img[i,j])
        lab_L[i, j] = (Lab[0] + value, Lab[1] + value, Lab[2] + value)
        lab_a[i, j] = (Lab[0], Lab[1] - value, Lab[2] + value) 
        lab_b[i, j] = (Lab[0], Lab[1] + value, Lab[2] + value) 
    both = np.hstack((img_new,lab_L,lab_a,lab_b))
    display(HTML("<font color='white'>|---------------Интенсивность--------------|-----------------L----------------|----------------a----------------|----------------b----------------|</font>"))
    cv2_imshow(both)

fSlider.observe(on_value_change, names='value')
display(fSlider)
```
